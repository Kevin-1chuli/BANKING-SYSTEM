```python
import os
import pandas as pd
```

```python
folder_path = "/content/drive/MyDrive/Colab Notebooks/datasets"
csv_file = "metadata.csv"

def load_data(folder_path, csv_path):
    csv_path = os.path.join(folder_path, csv_file)
    if not os.path.exists(csv_path):
      raise FileNotFoundError(f"{csv_file} not found. Make sure the file exists!")
    return pd.read_csv(csv_path)

meta_data = load_data(folder_path, csv_file)
meta_data.head()
```

Here i am structuring my grouping by zip_code!

```python
meta_data = meta_data.dropna(subset = ["zip_code"])
```

```python
meta_data.groupby("zip_code").size()
```

**Calculating the transformer capacity**

```python
meta_data['max_power'] = meta_data[['p1','p2','p3','p4','p5','p6']].max(axis=1)
capacity = meta_data['max_power'].sum() * 0.8
```
